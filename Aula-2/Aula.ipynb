Importação das Bibliotecas e Dados

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

fonte = "https://raw.githubusercontent.com/alura-cursos/imersao-dados-2-2020/master/MICRODADOS_ENEM_2019_SAMPLE_43278.csv"

dados = pd.read_csv(fonte)
dados.head()
```

```python
alunos_menor_quatorze = dados.query("NU_IDADE <= 14")
alunos_menor_quatorze["SG_UF_RESIDENCIA"].value_counts().plot.pie(figsize = (10, 8))
```

Gráfico de barras que mostra a distribuição dos canditatos com menos de 14 anos

```python
alunos_menor_quatorze["SG_UF_RESIDENCIA"].value_counts().plot.bar(figsize = (10, 8))
```

```python
plt.figure(figsize = (10,8))
sns.boxplot(x = "Q006", y = "NU_NOTA_MT", data = dados)
plt.title("Box Plot das Notas de Matemática pela Renda")
```

```python
renda_ordenada = dados["Q006"].unique()
renda_ordenada.sort()
print(renda_ordenada)
```

```python
plt.figure(figsize = (10,8))
sns.boxplot(x = "Q006", y = "NU_NOTA_MT", data = dados, order = renda_ordenada)
plt.title("Box Plot das Notas de Matemática pela Renda")
```

```python
provas = ["NU_NOTA_CN", "NU_NOTA_LC", "NU_NOTA_CH", "NU_NOTA_MT", "NU_NOTA_REDACAO"]
```

```python
dados[provas].sum(axis = 1)
```

```python
dados["NU_NOTA_TOTAL"] = dados[provas].sum(axis = 1)
```

```python
dados.head()
```

```python
plt.figure(figsize = (10,8))
sns.boxplot(x = "Q006", y = "NU_NOTA_TOTAL", data = dados, order = renda_ordenada)
plt.title("Box Plot das Notas Totais pela Renda")
```

```python
sns.displot(dados, x = "NU_NOTA_TOTAL")
```

```python
dados.query("NU_NOTA_TOTAL == 0")[provas]
```

```python
provas.append("NU_NOTA_TOTAL")
dados[provas].query("NU_NOTA_TOTAL == 0")
```

```python
dados_sem_nota_zero = dados.query("NU_NOTA_TOTAL != 0 ")
dados_sem_nota_zero.head()
```

```python
plt.figure(figsize = (10,8))
sns.boxplot(x = "Q006", y = "NU_NOTA_TOTAL", data = dados_sem_nota_zero, order = renda_ordenada)
plt.title("Box Plot das Notas Totais pela Renda")
```

```python
plt.figure(figsize = (10,8))
sns.boxplot(x = "Q006", y = "NU_NOTA_TOTAL", data = dados_sem_nota_zero, hue = "IN_TREINEIRO",order = renda_ordenada)
plt.title("Box Plot das Notas Totais pela Renda")
```
