### Inclusão das bibliotecas e base de dados

```python
import pandas as pd 
import matplotlib.pyplot as plt

fonte = "https://raw.githubusercontent.com/alura-cursos/imersao-dados-2-2020/master/MICRODADOS_ENEM_2019_SAMPLE_43278.csv"
dados = pd.read_csv(fonte)
```

<h3> Desafio 1 </h3>
<p> Mostrar a proporção de inscritos por idade </p>

```python
dados["NU_IDADE"].value_counts(normalize = True) * 100
```

<h3> Desafio 2 </h3>
<p> Descobrir de quais estados são os incritos com 13 anos </p>

```python
dados.query("NU_IDADE == 13")["SG_UF_RESIDENCIA"].unique()
```

<h3> Desafio 3 </h3>
<p> Colocar título no gráfico</p>

```python
dados["NU_IDADE"].hist(bins = 20, figsize = (8,6), color = "salmon").set_title("Distribuição das Idades")
```

<h3> Desafio 4 </h3>
<p> Plotar o Histograma das idades dos Treineiros e Não Treineiros </p>

```python
dados.query("IN_TREINEIRO == 0")["NU_IDADE"].hist(bins = 20, figsize = (8,6), color = "green").set_title("Idade dos Não Treineiros")
```

```python
dados.query("IN_TREINEIRO == 1")["NU_IDADE"].hist(bins = 20, figsize = (8,6), color = "green").set_title("Idade dos Treineiros")
```

<h3> Desafio 5 </h3>
    <p> Comparar as Distribuições da prova em Inglês e Espanhol </p>

```python
dados.query("TP_LINGUA == 0")["NU_NOTA_LC"].hist(bins = 20, figsize = (8,6), color = "gold")
dados.query("TP_LINGUA == 1")["NU_NOTA_LC"].hist(bins = 20, figsize = (8,6), color = "lightskyblue")

plt.title("Distribuição das notas em Linguagens e Códigos em Inglês e Espanhol")
plt.legend(["Inglês", "Espanhol"])
```
